#### 8.

```python
def redescendre(L, gauche):
    for i in range(len(L)-1, gauche, -1):
        if L[i] < L[i-1]:
            L[i], L[i-1] = L[i-1], L[i]
    gauche += 1
    return L, gauche

def trier(L):
    gauche = 0
    while gauche < len(L):
        L, gauche = redescendre(L, gauche)
    return L

trier([3,1,5,2,4])
```

#### 10.

```python
def redescendre(L, gauche, droite):
    for i in range(droite, gauche, -1):
        if L[i] < L[i-1]:
            L[i], L[i-1] = L[i-1], L[i]
    gauche += 1
    return L, gauche

def remonter(L, gauche, droite):
    for i in range(gauche, droite):
        if L[i] > L[i+1]:
            L[i], L[i+1] = L[i+1], L[i]
    return L, droite

def tri_boustrophedon(L):
    gauche, droite = 0, len(L)-1
    while gauche < droite :
        L, droite = remonter(L, gauche, droite)
        L, gauche = redescendre(L, gauche, droite)
    return(L)

tri_boustrophedon([6,8,3,5,4,2,1,9,7,0])
```

## Exercice 4

```python
class pile():
    def __init__(self):
        self.pile = []
    def fill(self, s, r):
        for i in range(r[1], r[0]-1, -1):
            self.append(s[i])
    def append(self, element):
        self.pile.append(element)
    def pop(self):
        assert not self.isEmpty()
        return self.pile.pop()
    def isEmpty(self):
        return self.pile == []
    def top(self):
        assert not self.isEmpty()
        return self.pile[-1]
```

#### 12.

```python
def scm(s):
    """Renvoie la liste ordonnée des couples d'indices correspondant au partitionnement en scm de s.
    s est de typle list."""
    assert type(s) == list
    
    s.append(None)
    rep = pile()
    li = 0 #abréviation de last index
    for i in range(len(s)-1):
        if s[i+1] is None:
            rep.append((li, i))
        elif s[i+1] <= s[i]:
            rep.append((li, i))
            li = i+1
    s.pop()
    return rep

scm([3,4,8,11,1,5,2,7,9,0,10,0]).pile
```

#### 13.

```python
def fusionner(s, r1, r2):
    """Fusionne les scm r1 et r2 en une seule scm.
    s est de type list.
    r1 et r2 sont deux scm consécutives de type tuple."""
    assert type(s) == list
    assert type(r1) == tuple
    assert type(r2) == tuple
    
    p1, p2 = pile(), pile()
    p1.fill(s, r1)
    p2.fill(s, r2)
    rep = []
    repi = [i for i in range(r1[0], r2[1]+1)]
    while not (p1.isEmpty() or p2.isEmpty()):
        if p1.top() < p2.top():
            rep.append(p1.pop())
        else:
            rep.append(p2.pop())
    while not p1.isEmpty():
        rep.append(p1.pop())
    while not p2.isEmpty():
        rep.append(p2.pop())
    for i in range(len(rep)):
        s[repi[i]] = rep[i]
    return(s)
        
fusionner([3,4,8,11,1,5,2,7,9,0,10,0], (0,3), (4,5))
```

#### 14.

```python
def depileFusionneRemplace(s, pile):
    """Retire les deux scm en haut de la pile, les fusionne dans la liste s, et replace les incices de la scm fusionnée au sommet de la pile.
    s est de type list.
    pile est une pile de scm."""
    assert type(s) == list
    
    r2 = pile.pop()
    r1 = pile.pop()
    s = fusionner(s, r1, r2)
    pile.append((r1[0], r2[1]))
    return s
    
s = [3,4,8,11,1,5,2,7,9,0,10,0]
p = pile()
p.append((0,3))
p.append((4,5))
print(depileFusionneRemplace(s, p), p.pile)
```

#### 15.

```python
def alphaTri(s):
    """Trie la liste s.
    s est une liste partiellement triée."""
    assert type(s) == list
    
    scm_p = scm(s)
    while not scm_p.top()[0] == 0:
        depileFusionneRemplace(s, scm_p)
    return s

alphaTri([3,4,8,11,1,5,2,7,9,0,10,0])
```
