## I. Définition

**def:**
Une <span class="mark">fonction récursive</span> est une fonction qui <span class="mark">s'appelle elle-même</span>. C'est la version informatique de la récurrence mathématique.

**Exemple:**
Si on définit une suite $u_n$ telle que
* $u_0 = 2$
* $u_n = \frac{1}{2}\left(u_{n-1} + \frac{3}{u_{n-1}}\right)$ $\forall n \in \mathbb{N}^{*}$

On peut la programmer comme ci-dessous:

```python
def u(n):
    if n == 0:
        return 2
    else:
        return 0.5*(u(n-1) + 3 / u(n-1))
    
u(0)
u(1)
u(2)
```

Ainsi $u_0$ renvoie directement 2, c'est la **<span class="mark">condition d'arrêt</span>**. Si n est supérieur à 0, la fonction s'appelle elle même avec un argument inférieur, qui s'appelle elle même avec un argument inférieur, etc.. jusqu'à atteindre la condition d'arrêt. **<span class="mark">Il ne faudra jamais oublier cette condition d'arrêt sans quoi le programme ne s'arrêterait plus</span>**.

On peut souvent programmer une même chose en utilisant une <span class="mark">fonction itérative</span> (avec une boucle) ou une <span class="mark">fonction récursive</span>.

Nous allons l'expérimenter avec la fonction factorielle qui s'y prête bien. C'est une fonction définie sur les entiers naturels telle que:
$$n! = \prod_{k = 1}^{n} k = n\times \prod_{k = 1}^{n-1} = n \times (n - 1)!$$

On peut donc calculer $n!$ en faisant une boucle. Une variable est initialisée à 1 puis est multipliée par un entier plus grand à chaque tout de boucle. On utilise en fait la définition $n! = \prod_{k = 1}^{n} k$. 

On peut aussi calculer $n!$ en utilisant la définition $n! = n \times (n - 1)!$ et la condition d'arrêt $1! = 1$.

### Exercice 2
1. Proposer une implémentation de la fonction factorielle en itératif
2. Même question en récursif.
3. Calculer $6!$ avec vos deux fonctions et comparer le résultat

```python
# question 1
def factorielle(n):
    """Calcule la factorielle de n de façon itérative."""
    assert type(n) == int, "n doit être un entier."
    assert n > 0, "n doit être positif et non nul."
    resultat = 1
    for i in range(n, 0, -1):
        resultat *= i
    return resultat
```

```python
# question 2
def f(n):
    """Calcule la factorielle de n de façon récurcive."""
    assert type(n) == int, "n doit être un entier."
    assert n > 0, "n doit être positif et non nul."
    if n == 1:
        return 1
    else:
        return n*f(n-1)
```

```python
# question 3
factorielle(4)
f(4)
```

Lorsque la définition de la fonction est récursive, il sera très facile de la programmer de cette manière. Idem lorsque la définition est itérative.

### Exercice 4
1. Proposer une fonction **puissance_iter** prenant comme argument un float **x** et n entier **n** et qui renvoie $x^n$ calculé de manière itérative.
2. Proposer une fonction **puissance_rec** prenant comme argument un float **x** et n entier **n** et qui renvoie $x^n$ calculé de manière récursive.

```python
# question 1
def puissance_iter(x, n):
    assert type(x) == int or type(x) == float, "x doit être un nombre."
    assert type(n) == int, "n doit être un entier."
    reponse = x
    if n > 0:
        for i in range(n-1):
            reponse *= x
    elif n < 0:
        for i in range(-1, n-2, -1):
            reponse /= x
    else:
        return 1
    return reponse
puissance_iter(2,-3)
```

```python
# question 2
def puissance_rec(x, n):
    assert type(x) == int or type(x) == float, "x doit être un nombre."
    assert type(n) == int, "n doit être un entier."
    if n == 0:
        return 1
    elif n > 0:
        return x*puissance_rec(x,n-1)
    else:
        return puissance_rec(x,n+1)/x
puissance_rec(2,-3)
```

Pour la version itérative, il faut utiliser $x^n = \prod_{k=1}^{n} x$


Pour la version recursive, on utilise $x^n = x \times x^{n-1}$ avec $x^0 = 1$
