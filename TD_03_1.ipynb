## Exercice 1. Crible d'Eratosthène

On choisit un entier N correspondant à la limite que l’on se fixe.
On va ensuite aligner tous les entiers (en barrant 0 et 1 avant de commencer) et effectuer le
processus suivant :
* Déterminer le plus petit entier n non barré et non encore étudié

* s’il reste des nombres non barrés > n, retourner à la première étape


A la fin, tous les nombres non barrés sont des nombres premiers.
Pour barrer des nombres, on va utiliser un tableau de booléens dont tous les éléments sont
initialement à True. On passera l’élément à False lorsque l’on veut barrer l’entier correspondant à
l’indice.

Nous avons déjà travaillé sur les algorithmes de calcul de puissance. Nous allons travailler ici sur la puissance modulaire.

2. Ecrire une fonction **puissance_mod_rapide** récursive qui prend en argument un float **x**, un entier positif ou nul **n** et un entier **m** et qui renvoie la valeur de $x^n\ mod\ m$ calculée par $x^0 = 1$ et $x^n = (x^{n//2}\ mod\ m) \times (x^{n//2}\ mod\ m)$ si n est pair et $x^n = (x\ mod\ m) \times (x^{n//2}\ mod\ m) \times (x^{n//2}\ mod\ m)$ si n est impair.

3. Expliquer ce qu'apporte la deuxième fonction par rapport à la première.

4. Utiliser le module **time** afin d'étudier l'évolution du temps nécessaire pour réaliser ces calculs avec les deux fonctions. Les résultats sont-il en accord avec la complexité de chacune de ces fonctions?

```python
# Question 1.
def puissance_mod(x, n, m):
    assert type(x) == float or type(x) == int
    assert type(n) == int and n >= 0
    assert type(m) == int
    if n == 0:
        return 1
    else:
        return (x%m)*(puissance_mod(x, n-1, m)%m)

puissance_mod(2, 5, 32)
```

```python
# Question 2.
def puissance_mod_rapide(x, n, m):
    assert type(x) == float or type(x) == int
    assert type(n) == int and n >= 0
    assert type(m) == int
    if n == 0:
        return 1
    else:
        temp = puissance_mod_rapide(x, n//2, m)
        if n%2 == 0:
            return (temp%m)*(temp%m)
        else:
            return (x%m)*(temp%m)*(temp%m)

puissance_mod_rapide(2, 5, 32)
```
