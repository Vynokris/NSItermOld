## III. La File


### II.1. Définition et propriétés



Il faut imaginer une **<span class="mark">file d’attente</span>**, premier arrivé premier servi!
ceci se note **queue** ou **FIFO** <span class="mark">first in first out</span> in english

Il faut pouvoir :
* créer une file vide => init
* tester si une file est vide => isEmpty
* ajouter un élément à la file => enQueue
* lire le premier élément de la file sans la modifier => first
* retirer le premier élément de la file et le renvoyer => deQueue
    
**Chacune de ces opérations se fait en <span class="mark">temps constant</span>.**

### Exercice 2:
Complétez ces 5 fonctions permettant d'obtenir une structure de **file** en <span class="mark">utilisant **deux piles</span>**. 
- Il faudra <span class="mark">ajouter les éléments dans la pile **instack</span>**
- Il faudra <span class="mark">lire ou défiler les éléments de la pile **outstack</span>**.
- Lorsque l'on a besoin de <span class="mark">**outstack**</span> et que celle ci est <span class="mark">vide, on retourne au préalable **instack** dans **outstack**</span>.


```python
class Stack():
    def __init__(self):
        self.data = []
    
    def isEmpty(self):
        return self.data == []
    
    def push(self,element):
        self.data.append(element)
    
    def top(self):
        assert not self.isEmpty()
        return self.data[-1]
    
    def pop(self):
        assert not self.isEmpty()
        return self.data.pop()
```

```python
def init():
    """Initialise une queue basée sur deux piles."""
    instack = Stack()
    outstack = Stack()
    return (instack,outstack)

def isEmpty(queue):
    """Retourne (True, True) si les deux piles qui composent la queue sont vide. Peut aussi retourner (False, False), (True, False) ou (False, True)."""
    return queue[0].isEmpty() and queue[1].isEmpty()

def enQueue(queue, element):
    """Ajoute l'élément en entrée dans la première pile qui constitue la queue."""
    queue[0].push(element)

def first(queue):
    """Renvoie le dernier élément de la deuxième pile qui constitue la queue. Si elle est vide, dépile la première pile dans la deuxième, puis renvoie le dernier élément de la deuxième pile."""
    assert not isEmpty(queue), "Impossible, la file est vide."
    if not queue[1].isEmpty():
        return queue[1].top()
    else:
        while not queue[0].isEmpty():
            queue[1].push(queue[0].pop())
        return queue[1].top()

def deQueue(queue):
    """Même fonctionnement que first(), mais dépile l'élément qu'elle renvoie."""
    assert not isEmpty(queue), "Impossible, la file est vide."
    if not queue[1].isEmpty():
        return queue[1].pop()
    else:
        while not queue[0].isEmpty():
            queue[1].push(queue[0].pop())
        return queue[1].pop()

q = init()
isEmpty(q)
enQueue(q, 2)
isEmpty(q)
enQueue(q, 9)
first(q)
enQueue(q, 6)
deQueue(q)
deQueue(q)
deQueue(q)
```
