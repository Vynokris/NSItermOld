## Exercice 1. Compter les arbres binaires

Un arbre binaire est un arbre pour lequel chaque sommet possède 0 ou 2 fils. Un <span class="mark">**noeud**</span> est un sommet qui possède <span class="mark">deux fils</span> et une <span class="mark">**feuille**</span> est un sommet <span class="mark">sans fils</span>. La figure ci-dessous donne trois exemples d'abres binaires pour 0, 1 et 4 noeuds.
![arbres](img/arbres_binaires.png)

1. Ecrire une fonction tree qui prend comme argument un entier **n** représentant le nombre de noeuds et qui renvoie un entier correspondant au nombre d'arbres binaires possibles à n noeuds.
2. Dessinez au brouillon tous les arbres possibles pour 3 et 4 noeuds et vérifiez que votre fonction renvoie bien le bon résultat.
3. Déterminez tree(100) -il faudra peut être améliorer la fonction écrite dans un premier temps-


```python
# question 1.
from time import time
def tree(n):
    assert type(n) is int, "n doit être un entier."
    if n <= 1:
        return 1
    else:
        num = 0
        for i in range(n):
            num += tree(n-1-i)*tree(i)
        return num

t = time()
tree(15)
print(time()-t)
```

```python
# question 3.
from time import time
tab = [1]
def tree2(n):
    global tab
    assert type(n) is int, "n doit être entier."
    assert n >= 0, "n doit être positif"
    if n >= len(tab):
        num = 0
        for i in range(n):
            num += tree2(n-1-i)*tree2(i)
        tab.append(num)
    return tab[n]

t = time()
tree2(15)
print(time()-t)
```

## Exercice 2. La fonction puissance

1. Ecrire une fonction **puissance** récursive qui prend en argument un float **x** et un entier positif ou nul **n** et qui renvoie la valeur de $x^n$ calculée par $x^0 = 1$ et $x^n = x \times x^{n-1}$

2. Ecrire une fonction **puissance_rapide** récursive qui prend en argument un float **x** et un entier positif ou nul **n** et qui renvoie la valeur de $x^n$ calculée par $x^0 = 1$ et $x^n = x^{n//2} \times x^{n//2}$ si n est pair et $x^n = x \times x^{n//2} \times x^{n//2}$ si n est impair.

3. Expliquer ce qu'apporte la deuxième fonction par rapport à la première.

```python
# Question 1.
from time import time
def puissance(x, n):
    """Renvoie x**n avec x un nombre entier ou décimal et n un nombre entier positif."""
    assert type(x) == float or type(x) == int, "x doit être un entier ou un nombre décimal."
    assert type(n) is int and n >= 0, "n doit être un entier positif."
    if n == 0:
        return 1
    else:
        return x*puissance(x,n-1)

t = time()
puissance(2,100)
print(time()-t)
```

```python
# Question 2.
def puissance_rapide(x, n):
    """Renvoie x**n avec x un nombre entier ou décimal et n un nombre entier positif."""
    assert type(x) == int or type(x) == float, "x doit être un entier ou un nombre décimal."
    assert type(n) == int and n >= 0, "n doit être un entier positif."
    if n == 0:
        return 1
    else:
        temp = puissance_rapide(x, n//2)
        return x**(n%2)*temp*temp

t = time()
puissance_rapide(2,100)
print(time()-t)
```

```python
def puissance_plus_rapide(x, n, a=1):
    """Renvoie x**n avec x un nombre entier ou décimal et n un nombre entier positif."""
    assert type(x) == int or type(x) == float, "x doit être un entier ou un nombre décimal."
    assert type(n) == int and n >= 0, "n doit être un entier positif."
    if n == 0:
        return a
    else:
        return puissance_plus_rapide(x**2,n//2,a*x**(n%2))

t = time()
puissance_plus_rapide(2,100)
print(time()-t)
```

## Exercice 4. Recherche Vs Recherche dichotomique

1. Ecrire une fonction **recherche** qui prend en argument une liste de mots triée, et un mot à rechercher à l'intérieur et qui doit renvoyer l'indice correspondant par balayage de la liste.
2. Même question dans le cas d'une recherche dichotomique.
3. Expliquer l'intérêt de la recherche dichotomique dans ce cas.
4. Serait-il efficace de lancer un algorithme de tri sur une longue liste puis effectuer une recherche dichotomique ou de lancer directement une recherche par balayage ? Justifier

```python
# question 1.
def recherche(liste, mot, i=0):
    """
    Renvoie l'indice du mot dans la liste de mots entrée. 
    liste doit être composée de un ou plusieurs str et être triée. 
    mot doit être un str.
    i ne doit pas être défini.
    """
    assert type(liste) == list, "La liste de mot doit être de type list et doit être triée."
    assert type(mot) == str, "Le mot doit être de type str."
    if len(liste) == 0:
        return None
    elif liste[0] == mot:
        return i
    else:
        return recherche(liste[1:], mot, i+1)

l = ["Abracadacbra", "Baliverne", "Cadeau", "Drapeau", "Echo", "Frites"]
recherche(l, "Echo")
```

```python
# question 2.
def recherche_dichotomique(liste, mot, i = 0, j = -1):
    """
    Renvoie l'indice du mot dans la liste de mots entrée. 
    liste doit être composée de un ou plusieurs str et être triée. 
    mot doit être un str.
    i et j ne doivent pas être définis.
    """
    assert type(liste) == list, "La liste de mot doit être de type list et doit être triée."
    assert type(mot) == str, "Le mot doit être de type str."
    if not mot in liste:
        return None
    if j == -1:
        j = len(liste)-1
    if i == j:
        return i
    moy_i_j = (i+j)//2
    if liste[moy_i_j] == mot:
        return moy_i_j
    elif liste[moy_i_j] < mot:
        return recherche_dichotomique(liste, mot, moy_i_j+1, j)
    else:
        return recherche_dichotomique(liste, mot, i, moy_i_j-1)

l = ["Abracadabra", "Baliverne", "Cadeau", "Drapeau", "Echo", "Frites"]
recherche_dichotomique(l, "Baliverne")
```
