# <span style="color:red;"><center>TD_03_3 - RSA 2/2 algorithme de chiffrement</center></span>

Le chiffrement RSA est nommé d'après les initiales de ses trois inventeurs Ronald Rivest, Adi Shamir et Leonard Adleman. C'est une méthode de chiffrement asymétrique inventée en 1977 et encore utilisée aujourd'hui pour le commerce électronique ou encore l'échange de données confidentielles sur internet.


Le TD se découpe en deux parties, une première pour mettre en places les différents outils nécessaires puis une seconde sur l'algorithme RSA spécifiquement.

Pour pouvoir réaliser ce chiffrement, il faut générer une clé publique à diffuser et une clé privée secrète. La clé publique permet à n'importe qui de crypter un message vous étant destiné, la clé privée permettant à vous seul de lire le message.

Pour générer ces clés, il faut:
* choisir deux nombres premiers distincts **p** et **q** assez grands. C'est la taille de ces nombres premiers qui rend le système sûr même avec des ordinateurs puissants.
* calculer $n = p \times q$
* calculer $\phi(n) = (p-1) \times (q-1)$
* choisir **e** entre max(p,q) et $\phi(n)$ tel que $PGCD(\phi(n), e) = 1$. Pour cela on prendra un nombre aléatoire dans l'intervale, on vérifie la condition sur le PGCD et si elle n'est pas vérifiée, on change de **e**
* calculer **d** entre max(p,q) et $\phi(n)$ tel que $e \times d \equiv 1 (mod\ \phi(n))$
* la clé publique est alors le couple **(e, n)** et la clé privée **(d, n)**. 


La sécurité vient du caractère disymétrique des opérations de multiplication (nécessaire pour décrypter lorsqu'on connait la clé privé) et de factorisation (nécessaire pour obtenir $\phi(n)$ et donc **d** lorsqu'on ne connait pas la clé privée. Des compétitions de factorisation sont organisées régulièrement avec des prix de plusieurs dizaines de milliers de dollar afin de se rendre compte de l'état des systèmes de factorisations.


Pour crypter un message **M**, il suffira de calculer $X = M^e\ (mod\ n)$ 

Pour décrypter **X**, il suffira de calculer $X^d\ (mod\ n) = M^{e d}\ (mod\ n) = M$
