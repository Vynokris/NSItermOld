Le principe de Kerckchoffs s'énonce:
>La sécurité d'un système de chiffrement ne doit reposer que sur la clé


Le système de chiffrement pour être considéré comme efficace et sûr doit donc pouvoir être connu, étudié, attaqué etc... et rester fiable tant que la clé reste secrète.

Le grand problème des méthodes de chiffrement présentées jusqu'ici vient de la nécessité de posséder la même clé pour l'emmeteur et le receveur. Il faut donc à un moment donné faire passer la clé de l'un à l'autre et c'est cette étape qui est critique enterme de sécurité.

    
    

Dans le cas d'un chiffrement à clé publique, la fonction de chiffrement et de déchiffrement ne sont pas symétriques. Ainsi, il est possible de diffuser largement une clé "publique" servant à crypter, il sera trop difficile de retrouver la clé "privée" de déchiffrement. La clé privée de déchiffrement restera secrète et sera conservée uniquement par la personne ayant généré les clés.

Le destinataire du message génère donc une clé publique servant à crypter qu'il va diffuser et une clé privée servant à décoder qu'il va conserver secrète. La seule clé qui va circuler entre l'emmeteur et le destinataire est cette clé publique qui n'est pas critique pour la sécurité du système de chiffrement.
