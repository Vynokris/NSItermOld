## II. La Pile


### II.1. Définition et propriétés

Il faut imaginer que les <span class="mark">données sont empilées</span> comme dans une <span class="mark">pile d'assiettes</span>.



La <span class="mark">première donnée arrivée sera en dessous</span> de la pile puis la suivante juste dessus etc…



A chaque instant, <span class="mark">on n’a accès directement qu’à l’assiette qui est au sommet de la pile</span>



On appelle cette structure une **pile** ou **stack** en anglais. On parle encore de **LIFO** pour <span class="mark">last in first out</span>


Pour la structure de pile il faut pouvoir:
* créer une pile vide (de taille définie) => init
* vérifier si la pile est vide => isEmpty
* empiler un élément => push
* dépiler un élément et le renvoyer => pop
    
**Chacune de ces opérations se fait en <span class="mark">temps constant</span>.**

### Exercice 2:
Programmer ces 4 fonctions dans le cas d'une pile bornée, c'est à dire de taille définie à l'avance que l'on ne peut pas dépasser.


```python
def init(n):
    """Initialisation d'une pile de taille n"""
    return [0]+[None]*n

def isEmpty(stack):
    return stack[0] == 0

def push(element, stack):
    assert stack[0] < len(stack)-1, "La pile est pleine."
    stack[0] += 1
    stack[stack[0]] = element

def pop(stack):
    assert stack[0] != 0, "La pile est déjà vide."
    val = stack[stack[0]]
    stack[0] -= 1
    return val

a = init(2)
print(isEmpty(a))
push(1, a)
print(isEmpty(a))
push(2, a)
try:
    push(3, a)
except:
    print("La pile est pleine.")
pop(a)
pop(a)
print(isEmpty(a))
pop(a)
```
